### Finding Distribution of total_bill with respect to categorical column

```python
plt.figure(figsize=(16,5))
plt.subplot(1,4,1)
sns.histplot(data=df, x='total_bill', kde=True, hue='sex')
plt.title('bill - sex')

plt.subplot(1,4,2)
sns.histplot(data=df, x='total_bill', kde=True, hue='smoker')
plt.title('bill - smoker')

plt.subplot(1,4,3)
sns.histplot(data=df, x='total_bill', kde=True, hue='day')
plt.title('bill - day')

plt.subplot(1,4,4)
sns.histplot(data=df, x='total_bill', kde=True, hue='time')
plt.title('bill - time')
plt.show()
```

## Insights
- Male are supposed eat more food for around 10 to 25 dollars
- Non-smokers are supposed to eat more 
- People on saturday supposed to eat more food for around 20 dollars
- Dinner time is best for people to eat more food

### Finding Distribution of tip with respect to categorical columns

```python
plt.figure(figsize=(16,5))
plt.subplot(1,4,1)
sns.histplot(data=df, x='tip', kde=True, hue='sex')
plt.title('tip - sex')

plt.subplot(1,4,2)
sns.histplot(data=df, x='tip', kde=True, hue='smoker')
plt.title('tip - smoker')

plt.subplot(1,4,3)
sns.histplot(data=df, x='tip', kde=True, hue='day')
plt.title('tip - day')

plt.subplot(1,4,4)
sns.histplot(data=df, x='tip', kde=True, hue='time')
plt.title('tip - time')
plt.show()
```

## Insights
- Males are supposed to give more tips for around 3 dollars
- Non-Smoker gives more tips 
- On Saturday more tips were give for around 3 dollars and on Sunday the tips are high
- On Dinner time the tips count are high

### Analysing Size of Male and Female based on Smoker, Day and Time

```python
plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
sns.barplot(data=df, x='sex', y='size', hue='smoker')
plt.title('Size - Smoker')

plt.subplot(1,4,2)
sns.barplot(data=df, x='sex', y='size', hue='day')
plt.title('Size - Day')

plt.subplot(1,4,3)
sns.barplot(data=df, x='sex', y='size', hue='time')
plt.title('Size - Time')
```

## Insights
- Non-Smoker Male and Females are supposed to come higher
- On Sunday the Male and Female are higher means the sales are high on Sunday
- Males are supposed to come alot on Dinner but Females are higher on both

# Analyzing Total_bill for Males Only 

```python
plt.figure(figsize=(16,5))

plt.subplot(1,4,1)
sns.histplot(data=df[df.sex == 'Male'], x='total_bill', kde=True, hue='smoker')
plt.title('Male - total bill - smoker')

plt.subplot(1,4,2)
sns.histplot(data=df[df.sex == 'Male'], x='total_bill', kde=True, hue='day')
plt.title('Male - total bill - day')

plt.subplot(1,4,3)
sns.histplot(data=df[df.sex == 'Male'], x='total_bill', kde=True, hue='time')
plt.title('Male - total bill - time')
plt.show()
```

# Insights
- Non-Smokers Male are suppoed to eat more of range 15 to 20 dollars
- Males like to eat more on Saturday and Sunday
- Males are not available much at lunch but on Dinner Males are high

### Analysing bill of Females based on Smoker, Day, Time

```python
plt.figure(figsize=(16,5))

plt.subplot(1,4,1)
sns.histplot(data=df[df.sex == 'Female'], x='total_bill', kde=True, hue='smoker')
plt.title('Female - total bill - smoker')

plt.subplot(1,4,2)
sns.histplot(data=df[df.sex == 'Female'], x='total_bill', kde=True, hue='day')
plt.title('Female - total bill - day')

plt.subplot(1,4,3)
sns.histplot(data=df[df.sex == 'Female'], x='total_bill', kde=True, hue='time')
plt.title('Female - total bill - time')
plt.show()
```

# Insights 
- Non-Smoker Females are more likely to eat for around 15 dollars
- Females are more present on Thursday and Saturday
- On Lunch and Dinner Females quantity are almost same

### Visualizing the relation between total_bill and tip

```python
sns.scatterplot(data=df, x='total_bill', y='tip')
```

### Visualizing Relation between total_bill and tip based on sex, smoker, day, time

```python
plt.figure(figsize=(20,5))
plt.subplot(1,4,1)
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex')
plt.title('Total bill vs Tips - Sex')

plt.subplot(1,4,2)
sns.scatterplot(data=df, x='total_bill', y='tip', hue='smoker')
plt.title('Total bill vs Tips - Smoker')

plt.subplot(1,4,3)
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day')
plt.title('Total bill vs Tips - Day')

plt.subplot(1,4,4)
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time')
plt.title('Total bill vs Tips - Time')
```

# Insights
- For High bill Males are supposed to give high tips
- Bill was increasing on Saturday and Sunday and tips are also increasing
- Dinner will also increasing bills and tips

### Visualizing Counts of Male and Females based on Smoker, Dat, Time

```python
plt.figure(figsize=(13,5))
plt.subplot(1,4,1)
sns.countplot(data=df, x='sex', hue='smoker')
plt.title('Count Sex - Smoker')

plt.subplot(1,4,2)
sns.countplot(data=df, x='sex', hue='day')
plt.title('Count Sex - Day')

plt.subplot(1,4,3)
sns.countplot(data=df, x='sex', hue='time')
plt.title('Count Sex - Time')
```

# Insights
- Non-Smoker Males and Females are more likely to come
- Males are more likely to come on Saturday and Sunday / Females are more likely to come at Thursday and Saturday
- Males are Females are more likely to come at Dinner rather than Lunch
